```python
qrels = pd.read_csv(qrels_path,names=["qid","ignore","docId","relevance"],usecols=[0,2,3],delimiter=' ')
ret = pd.read_csv(retrieved_path,delimiter=' ')
ret.head(5)
```

```python
ret["relevance"] = np.zeros(ret.shape[0])
```

```python
def clean_qrels(df):
    df = df.astype(int)
    return df
qrels = clean_qrels(qrels)
```

## join qrels with ret

```python
for idx,line in enumerate(range(ret.shape[0])):
    shape_qrels = qrels[(qrels.docId == ret.iloc[idx].docId) & (qrels.qid == ret.iloc[idx].qid)].values.shape[0]
    if (shape_qrels == 1):
        ret.at[idx,"relevance"] = qrels[(qrels.docId == ret.iloc[idx].docId) & (qrels.qid == ret.iloc[idx].qid)].relevance.values[0]
    if (idx == 100 or idx == 500 or idx == 1000 or idx == 2000 or idx == 3000):
        print(idx)
ret = ret.fillna(0)
```

```python
ret.head(15)
```

```python
temp = ret[["docId", "qid", "relevance","BM_title_summary"]]
```

## generate file for ranklib

```python
def gen_csv(df,filepath):
    finalToCsv = df.copy()
    
    finalToCsv = finalToCsv.sort_values("qid",axis=0)
    
    finalToCsv.docId = finalToCsv.docId.apply(lambda x: "#" + str(x))
    finalToCsv.qid = finalToCsv.qid.apply(lambda x: "qid:" + str(x))

    features = [col for col in finalToCsv.columns if col != "docId" and col != "qid" and col != "relevance"]
    print(features)
    for idx,col in enumerate(features):
        if col != "docId" and col != "qid" and col != "relevance":
            finalToCsv[col] = finalToCsv[col].apply(lambda x: str(idx+1) + ':' + str(x))
    
    if not os.path.exists(filepath):
        os.makedirs(filepath)
    finalToCsv.to_csv(os.path.join(filepath,"rankLib.txt"),sep=' ',columns=["relevance","qid"] + features + ["docId"],index=False,header=False)
    return finalToCsv

finalToCsv = gen_csv(temp,working_dir)
```

## RankLib - com o script gerado, estão aqui os comandos a correr para treinar e obter resultados com o RankLib

```python
test = os.path.join(working_dir,"test.txt")
final = os.path.join(working_dir,"final.txt")
ranklib = os.path.join(working_dir,"ranklib.txt")
results_trec = os.path.join(working_dir,"trec_output.txt")
command_train = "java -jar {} -train {} -ranker 4 -metric2t MAP -norm linear -save {}".format(ranklib_path,ranklib,test)
command_test = "java -jar {} -load {} -metric2T MAP -norm linear -rank {} -indri {}".format(ranklib_path,test,ranklib,final)
command_trec = "{} {} {} > {}".format(path_trec_exe,qrels_path,final,results_trec)
```

```python
os.system(command_train) # train ranklib
os.system(command_test) # gen file with ranklib for trec eval input
os.system(command_trec) # gen trec output with new retrival
```
